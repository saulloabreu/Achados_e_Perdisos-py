Acessando dados da tabela Itens


```python
import sqlite3

DATABASE = 'DATABASES/0000_banco_dados.db'

def consultar_tabela(items):
    with sqlite3.connect(DATABASE) as conn:
        c = conn.cursor()
        c.execute(f"SELECT * FROM {items}")
        rows = c.fetchall()
        for row in rows:
            print(row)

# Testando com a tabela 'items'
consultar_tabela('items')

```

```python
import sqlite3
import pandas as pd

DATABASE = 'DATABASES/0000_banco_dados.db'

def consultar_tabela_com_pandas(items):
    with sqlite3.connect(DATABASE) as conn:
        df = pd.read_sql_query(f"SELECT * FROM {items}", conn)
    return df

# Testando com a tabela 'items'
df_items = consultar_tabela_com_pandas('items')

# Exibir o DataFrame
df_items

```
